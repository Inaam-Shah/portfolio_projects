```python
# Best season for running--> Slower in summer than winter?
# Spring 3-5
# Summer 6-8
# Fall 9-11
# Winter 12-2
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head(25)
```

```python
# Create a season column
# Spring 3-5
# Summer 6-8
# Fall 9-11
# Winter 12-2
#if (usa_races['race_season'] == 12 or usa_races['race_season'] == 1 or usa_races['race_season'] == 2):
# return "winter"
#elif(usa_races['race_season'] == 3 or usa_races['race_season'] == 4 or usa_races['race_season'] == 5):
# return "spring"
#elif(usa_races['race_season'] == 6 or usa_races['race_season'] == 7 or pusa_races['race_season'] == 8):
# return "summer"
#else:
# return "fall" 
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'winter' if x > 11 else 'fall' if x > 8 else 'summer' if x > 5 else 'spring' if x > 2 else 'winter')
```
